```python
from urllib.request import urlopen as ureq
```

```python
from bs4 import BeautifulSoup as soup
```

```python
myurl = 'https://www.newegg.com/global/in-en/Video-Cards-Video-Devices/Category/ID-38'
```

```python
uclient = ureq(myurl)
page_html = uclient.read()
uclient.close()
```

```python
page_soup = soup(page_html,"html.parser")
```

```python

page_soup
```

```python
l = []

containers = page_soup.findAll('div',{'class':'item-container'})
for container in containers:
    
    brand = container.div.div.a.img['title']
    title_container = container.findAll('a',{'class':'item-title'})
    product_name = title_container[0].text
    shiping_container = container.findAll('li',{'class':'price-current'})
    shipping = shiping_container[0].text.replace('(4 Offers)',"").replace('(3 Offers)',"").replace('(2 Offers)',"").replace('(5 Offers)',"").replace('-',"").strip()
    
    
    print(brand)
    print(product_name)
    print(shipping)
    print(" ")
```
